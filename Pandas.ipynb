```python
# Trier les valeurs selon l'identifiant par ordre croissant :
data_json.sort_values('identifiant')
```

```python
# Trier les valeurs selon l'identifiant par ordre décroissant : 
data_json.sort_values('identifiant', ascending=False)
```

```python
# Trier les valeurs selon le genre puis le nom, par ordre croissant 
data_json.sort_values(['genre', 'nom'])
```

```python
# Trier par genre en ordre croissant et en nom par ordre décroissant
data_json.sort_values(['genre', 'nom'], ascending=[True, False])
```
